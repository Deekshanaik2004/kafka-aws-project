```python
import requests
import pandas as pd
import schedule
import time
from datetime import datetime
import os

API_KEY = "cc9c8eb1296588f95055cc7cc31a65c7"

lat =15.3004543
lon = 74.0855134

def collect_weather():

    url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_KEY}&units=metric"

    response = requests.get(url)

    data = response.json()

    weather_data = {
        "time": datetime.now(),
        "temperature": data["main"]["temp"],
        "humidity": data["main"]["humidity"],
        "pressure": data["main"]["pressure"],
        "weather": data["weather"][0]["description"]
    }

    df = pd.DataFrame([weather_data])

    file_exists = os.path.isfile("weather_dataset.csv")

    df.to_csv(
        "weather_dataset.csv",
        mode="a",
        header=not file_exists,
        index=False
    )

    print("Weather data collected:", weather_data)

schedule.every(1).minutes.do(collect_weather)

print("Weather collector started...")

while True:
    schedule.run_pending()
    time.sleep(1)
```
